```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from astropy.cosmology import LambdaCDM
from astropy.cosmology import Planck15
import astropy.units as u

z_list = []
z_test = []

for i in range(50):
    z_list.append(0.006 + i*0.01)


z_test = np.array(z_list)

H0_new=67.7

# Planck model - flat with Lambda

v_flat_wL = []
dist_flat_wL = []
comov_flat_wL= []
hz_flat_wL = []

cosmo_flat_wL = LambdaCDM(H0=H0_new, Om0=0.2, Ode0=0.8)

#dist_planck=Planck15.angular_diameter_distance(z_test)
L_dist_flat_wL=cosmo_flat_wL.luminosity_distance(z_test)
DM_flat_wL = 5*np.log10(L_dist_flat_wL.value * 1e6/10)

A_dist_flat_wL = cosmo_flat_wL.angular_diameter_distance(z_test)
# print(DM_planck)

comov_flat_wL = cosmo_flat_wL.comoving_distance(z_test)
hz_flat_wL = cosmo_flat_wL.H(z_test)
v_flat_wL = hz_flat_wL*comov_flat_wL/(1+z_test)


# Open no Lambda model
v_open = []
dist_open = []
comov_open = []
hz_open = []

cosmo_open = LambdaCDM(H0=H0_new, Om0=0.2, Ode0=0)

#dist_open=cosmo_open.angular_diameter_distance(z_test)
L_dist_open=cosmo_open.luminosity_distance(z_test)
DM_open = 5*np.log10(L_dist_open.value * 1e6/10)

A_dist_open = cosmo_open.angular_diameter_distance(z_test)

comov_open = cosmo_open.comoving_distance(z_test)
hz_open = cosmo_open.H(z_test)
v_open = hz_open*comov_open/(1+z_test)


# Flat no Lambda model

v_flat_noL = []
dist_flat_noL = []
comov_flat_noL = []
hz_flat_noL = []

# cosmo_flat = LambdaCDM(H0=67.7, Om0=0.307, Ode0=0.693)
cosmo_flat_noL = LambdaCDM(H0=H0_new, Om0=1, Ode0=0)

# dist_flat=cosmo_flat.angular_diameter_distance(z_test)
L_dist_flat_noL=cosmo_flat_noL.luminosity_distance(z_test)
DM_flat_noL = 5*np.log10(L_dist_flat_noL.value * 1e6/10)

A_dist_flat_noL = cosmo_flat_noL.angular_diameter_distance(z_test)


comov_flat_noL = cosmo_flat_noL.comoving_distance(z_test)
hz_flat_noL = cosmo_flat_noL.H(z_test)
v_flat_noL = hz_flat_noL*comov_flat_noL/(1+z_test)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=z_test, y=DM_flat_wL, name="Flat, with Lambda"))
fig.add_trace(go.Scatter(x=z_test, y=DM_open, name="Open"))
fig.add_trace(go.Scatter(x=z_test, y=DM_flat_noL, name="Flat, no Lambda"))

fig.update_xaxes(type="log")

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=z_test, y=L_dist_flat_wL, name="Flat, with Lambda"))
fig.add_trace(go.Scatter(x=z_test, y=L_dist_open, name="Open"))
fig.add_trace(go.Scatter(x=z_test, y=L_dist_flat_noL, name="Flat, no Lambda"))


#fig.update_xaxes(type="log")
#fig.update_yaxes(type="log")

fig.show()
```

```python
fig = go.Figure()

diff_flat_wL = DM_flat_wL - DM_open
diff_open = DM_open - DM_open
diff_flat_noL = DM_flat_noL - DM_open

fig.add_trace(go.Scatter(x=z_test, y=diff_flat_wL, name="Flat with Lambda"))
fig.add_trace(go.Scatter(x=z_test, y=diff_open, name="Open"))
fig.add_trace(go.Scatter(x=z_test, y=diff_flat_noL, name="Flat, no Lambda"))

fig.update_xaxes(type="log")

fig.show()
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(x=L_dist_flat_wL, y=v_flat_wL, name="Flat with Lambda"))
fig.add_trace(go.Scatter(x=L_dist_open, y=v_open, name="Open"))
fig.add_trace(go.Scatter(x=L_dist_flat_noL, y=v_flat_noL, name="Flat no Lambda"))

#fig.update_xaxes(type="log")

fig.show()
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(x=A_dist_flat_wL, y=v_flat_wL, name="Flat with Lambda"))
fig.add_trace(go.Scatter(x=A_dist_open, y=v_open, name="Open"))
fig.add_trace(go.Scatter(x=A_dist_flat_noL, y=v_flat_noL, name="Flat no Lambda"))

#fig.update_xaxes(type="log")
#fig.update_yaxes(type="log")

fig.show()
```
